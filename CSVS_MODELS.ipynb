```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 
```

```python
df=vertical_concat[['x','y','z','label']]
X = df.drop("label", axis = 1)
y = df['label']
print('Shape of X = ', X.shape)
print('Shape of y = ', y.shape)
d={'Running':0,'Walking':1, 'Standing':2,'Not with me':3,'Phone in use':4} # Not with me"Phone in use"Running"Walking"Standing
y=df['label'].map(d)
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.2, random_state=51)
print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
clf1=KNeighborsClassifier(7)
clf1=clf1.fit(X_train, y_train)
pred=clf1.predict(X_test)
print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
print("Report : \n", classification_report(y_test, pred))
print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
```

```python
#Storing the model
import pickle
filename = 'KNN.sav'
pickle.dump(clf1, open(filename, 'wb'))
loaded_model1 = pickle.load(open(filename, 'rb'))
```
